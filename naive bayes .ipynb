```python
from xlrd import open_workbook
import math

wb = open_workbook('data.xlsx')
sheet = wb.sheet_by_index(0)
 
route = []
vaicle = []
service = []

for j in range(1,sheet.nrows):
    route.append( int(sheet.cell_value(j,0)))  

for j in range(1,sheet.nrows):
    vaicle.append( int(sheet.cell_value(j,2)))       

for j in range(1,sheet.nrows):
    service.append( int(sheet.cell_value(j,1))) 

#for j in range(1,len(service)):
#    print(service[j]) 
```

```python
total = sheet.nrows-1
p_uber   = service.count(1)/total
p_pathoe = service.count(2)/total
p_shohoz = service.count(3)/total
c11 = c12 = c13 = 0
p_route = []

 
for i in range(1, max(route)+1):
    for j in range(0, len(service)):
        if(route[j] == i):
            if(service[j] == 1):
                c11 +=1
            elif(service[j] == 2):
                c12 +=1
            else:
                c13 +=1
        
    #print(c11," ",c12," ",c13)
    p_route.append(c11/service.count(1))
    p_route.append(c12/service.count(2))
    p_route.append(c13/service.count(3))
    c11 = c12 = c13 = 0

    
#for k in range(0,len(p_route)):
#    print(p_route[k])
    
#print(route.count(3))  
#print(max(route))
```

```python
p_vaicle = []

for i in range(1, max(vaicle)+1):
    for j in range(0, len(service)):
        #print(j)
        if(vaicle[j] == i):
            if(service[j] == 1):
                c11 +=1
            elif(service[j] == 2):
                c12 +=1
            else:
                c13 +=1
        
    #print(c11," ",c12," ",c13)
    p_vaicle.append(c11/service.count(1))
    p_vaicle.append(c12/service.count(1))
    p_vaicle.append(c13/service.count(1))
    c11 = c12 = c13 = 0
    
    
#for k in range(0,len(p_vaicle)):
#    print(p_vaicle[k])
```

```python
route_in= input()
vaicle_in=input()
i = j  = 0
c=0
#print(route_in)
#print(vaicle_in)



if(int(route_in) == 1):     
    i = 0
elif(int(route_in) == 2):
    i = 3
elif(int(route_in)== 3):
    i = 6
elif(int(route_in) == 4):
    i = 9

    
#print(i)

if(int(vaicle_in) == 1):
    j = 0
elif(int(vaicle_in) == 2):
    j = 3
elif(int(vaicle_in) == 3):
    j = 6


t_uber = p_route[i] * p_vaicle[j] 
t_pathoe = p_route[i+1] * p_vaicle[j+1] 
t_shohoz = p_route[i+2] * p_vaicle[j+2] 


likelyhood = (t_uber * p_uber) + (t_pathoe * p_pathoe) + (t_shohoz * p_shohoz)
#print(likelyhood)
#print(t_uber," ",t_pathoe," ",t_shohoz)    

uber_class = (t_uber * p_uber)/likelyhood
pathoe_class = (t_pathoe * p_pathoe)/likelyhood
shohoz_class = (t_shohoz * p_shohoz)/likelyhood

if(uber_class >= pathoe_class and uber_class >= shohoz_class):
    print("***Uber*** is best choise")
elif(pathoe_class >= uber_class and pathoe_class >= shohoz_class):
    print("***Pathoe*** is best choise")
elif(shohoz_class >= pathoe_class and shohoz_class >= uber_class):
    print("***Shohoz*** is best choise")


    
print(uber_class," ",pathoe_class," ",shohoz_class)
```
