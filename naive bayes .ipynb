```python
from xlrd import open_workbook
import math

wb = open_workbook('data_final.xlsx')
sheet = wb.sheet_by_index(0)


route = []
vaicle = []
service = []
behave = []
available = []
traffic = []
experience = []
gender = []


for j in range(1,sheet.nrows):
    route.append(int(sheet.cell_value(j,0)))

for j in range(1,sheet.nrows):
    vaicle.append( int(sheet.cell_value(j,2)))

for j in range(1,sheet.nrows):
    service.append( int(sheet.cell_value(j,1))) 


for j in range(1,sheet.nrows):
    behave.append( int(sheet.cell_value(j,3))) 
    
for j in range(1,sheet.nrows):
    available.append( int(sheet.cell_value(j,4)))
    
for j in range(1,sheet.nrows):
    traffic.append( int(sheet.cell_value(j,5)))
    
for j in range(1,sheet.nrows):
    experience.append( int(sheet.cell_value(j,6)))
    
for j in range(1,sheet.nrows):
    gender.append( int(sheet.cell_value(j,7)))
    
#for j in range(0,len(gender)):
#    print(gender[j]) 
```
